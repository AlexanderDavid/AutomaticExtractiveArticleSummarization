```python
import matplotlib.pyplot as plt
import numpy as np
from math import log
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
N = 500

def idf(x: int) -> float:
  return log(N / x)

def idf_smooth(x: int) -> float:
  return log(1 + (N / x))

def idf_probabilistic(x: int) -> float:
  return log((N - x)/x)
```

```python
# Data
df=pd.DataFrame({'x': range(1,500),
                 'Inverse Document Frequency': [idf(x) for x in range(1, 500)],
                 'Inverse Document Frequency Smoothed': [idf_smooth(x) for x in range(1, 500)],
                 'Probabilistic Inverse Document Frequency': [idf_probabilistic(x) for x in range(1, 500)] })
```

```python
# Set the font to be serif, rather than sans
sns.set(style="whitegrid", font="Serif")
```

```python
sns.set_context("paper")
sns_plot = sns.lineplot(data=df, size=10)
sns_plot.set_ylabel("Inverse Document Score")
sns_plot.set_xlabel("Frequency of Documents Containing the Term")
sns_figure = sns_plot.get_figure()
sns_figure.savefig("plots/idf_comparison.png")
```
